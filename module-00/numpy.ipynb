## 3. Creating NumPy Arrays
NumPy arrays are the core data structure of the library. They can be created from Python lists or using built-in functions.

### 3.1. From Python Lists

```python
# 1D array.
arr_1d = np.array([1, 2, 3])

# 2D array (matrix).
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
```

### 3.2. Using Built-in Functions

```python
# Zeros array.
zeros_arr = np.zeros((2, 3))  # 2x3 array of zeros

# Ones array.
ones_arr = np.ones((3, 2))  # 3x2 array of ones

# Identity matrix.
identity_mat = np.eye(4)  # 4x4 identity matrix
```

## 4. Array Operations
NumPy supports various array operations, including element-wise operations, matrix multiplication, and aggregation functions.

### 4.1. Element-wise Operations

```python
arr = np.array([1, 2, 3])

# Element-wise operations.
arr_squared = arr**2
arr_sum = arr + 5
```

### 4.2. Matrix Multiplication

```python
matrix_a = np.array([[1, 2], [3, 4]])
matrix_b = np.array([[5, 6], [7, 8]])

# Matrix multiplication.
result_matrix = np.dot(matrix_a, matrix_b)
```

### 4.3. Aggregation Functions

```python
arr = np.array([1, 2, 3, 4, 5])

# Aggregation functions.
mean_value = np.mean(arr)
sum_value = np.sum(arr)
max_value = np.max(arr)
```

## 5. Indexing and Slicing
NumPy arrays support advanced indexing and slicing operations, similar to Python lists.

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])

# Indexing.
element = arr[0, 1]  # Access the element at the first row and second column

# Slicing.
subarray = arr[:, 1:]  # Slice all rows, starting from the second column
```

## 6. Random Module in NumPy
NumPy provides a `random` module for generating random numbers and arrays.

```python
random_arr = np.random.rand(2, 3)  # 2x3 array of random numbers between 0 and 1
```

## 7. Broadcasting
NumPy supports broadcasting, a powerful mechanism that allows operations on arrays of different shapes and sizes.

```python
arr_a = np.array([1, 2, 3])
arr_b = np.array([[4], [5], [6]])

result = arr_a + arr_b
```

## 8. Array Shape and Reshaping
NumPy provides functions to retrieve the shape of an array and to reshape an array into a different shape.

### 8.1. Array Shape
The `shape` attribute of a NumPy array returns a tuple representing the array's dimensions.

```python
import numpy as np

# Creating a 2D array.
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])

# Getting the shape of the array.
print("Shape of arr_2d:", arr_2d.shape)
```

In this example, `arr_2d` is a 2D array with a shape of (2, 3), indicating 2 rows and 3 columns.

### 8.2. Reshaping Arrays
The `reshape` function allows you to change the shape of an array while maintaining the same data.

```python
import numpy as np

# Creating a 1D array.
arr_1d = np.array([1, 2, 3, 4, 5, 6])

# Reshaping the array to a 2x3 matrix.
arr_reshaped = arr_1d.reshape(2, 3)

print("Original array:")
print(arr_1d)
print("\nReshaped array:")
print(arr_reshaped)
```

In this example, `arr_1d` is reshaped into a 2x3 matrix (`arr_reshaped`). The total number of elements must remain the same after reshaping.

### 8.3. Flattening Arrays
Conversely, you can use the `flatten` method to transform a multi-dimensional array into a 1D array.

```python
import numpy as np

# Creating a 2x3 array.
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])

# Flattening the array.
arr_flattened = arr_2d.flatten()

print("Original 2D array:")
print(arr_2d)
print("\nFlattened array:")
print(arr_flattened)
```

The `flatten` method is useful when you need to convert a multi-dimensional array into a 1D array.

## 9. Linear Algebra Operations
NumPy includes a `linalg` module for linear algebra operations.

```python
matrix = np.array([[1, 2], [3, 4]])

# Matrix determinant.
det = np.linalg.det(matrix)

# Matrix inverse.
inverse_matrix = np.linalg.inv(matrix)
```

## 10. Working with Images in NumPy

NumPy can be utilized to load, manipulate, and analyze images by representing them as ndarrays.

### 10.1. Loading an Image
You can use the `matplotlib` library to load and display images. Ensure you have the library installed:

```python
import matplotlib.pyplot as plt
from matplotlib.image import imread

# Load an image into a NumPy ndarray.
image_path = 'usi.jpeg'  # Replace with the path to your image
img = imread(image_path)

# Create a copy of the image to allow modification.
img_copy = np.copy(img)

# Display the original image.
plt.imshow(img_copy)
plt.title('Original Image')
plt.show()
```

### 10.2. Getting Image Size
The shape attribute of the ndarray can be used to retrieve the dimensions of the image, representing the height, width, and color channels.

```python
# Get the size of the image.
height, width, channels = img_copy.shape
print(f"Image Size: {width}x{height} pixels with {channels} channels")
```

### 10.3. Understanding RGB
In a typical RGB image, each pixel is represented by three values: Red, Green, and Blue. These values range from 0 to 255, indicating the intensity of each color channel.

```python
# Accessing RGB values of a pixel.
pixel_value = img_copy[50, 100]
print("RGB Values at (50, 100):", pixel_value)
```

### 10.4. Changing Pixel Values
You can modify the values of individual pixels in the image.

```python
# Change the color of a specific pixel.
img_copy[50, 100] = [255, 0, 0]  # Set the pixel at (50, 100) to pure red
```

### 10.5. Displaying the Modified Image
Finally, display the modified image after changing pixel values:

```python
# Display the modified image
plt.imshow(img_copy)
plt.title('Modified Image')
plt.show()
```

### 10.6. Changing Pixel Values
You can modify the values of ranges of pixels in the image.

```python
# Change the color of a range of pixels.
img_copy[10:250, 30:750] = [0, 255, 0]  # Set a rectangular region to pure green

# Display the modified image
plt.imshow(img_copy)
plt.title('Modified Image')
plt.show()
```

### 10.7. Changing Lower Pixel Values
You can modify the values of low value pixels in the image.

```python
# Check pixels that are less than 200.
low_value_filter = img_copy < 200
print("Shape of low_value_filter:", low_value_filter.shape)

# Make those pixels equal to 0.
img_copy[low_value_filter] = 0

# Display the modified image
plt.imshow(img_copy)
plt.title('Modified Image')
plt.show()
```
